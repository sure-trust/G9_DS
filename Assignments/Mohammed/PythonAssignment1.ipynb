```python
# Hangman Game

def hangman():
    # List of words for the game
    words = ['python', 'hangman', 'computer', 'programming', 'developer', 'debugging']

    # Choose a random word from the list
    import random
    word = random.choice(words)

    # Initialize variables
    guessed_letters = []
    attempts = 6  # Number of attempts allowed

    print("Welcome to Hangman!")
    print("Try to guess the word.")

    # Display initial underscores representing each letter of the word
    guessed_word = ['_'] * len(word)
    print(' '.join(guessed_word))

    while attempts > 0:
        guess = input("Enter a letter: ").lower()

        # Check if the guess is a single letter
        if len(guess) != 1:
            print("Please enter only one letter.")
            continue

        # Check if the guess is a valid letter
        if not guess.isalpha():
            print("Please enter a valid letter.")
            continue

        # Check if the letter has already been guessed
        if guess in guessed_letters:
            print("You've already guessed that letter.")
            continue

        guessed_letters.append(guess)

        # Check if the guessed letter is in the word
        if guess in word:
            print("Correct!")
            # Update guessed_word with the correctly guessed letter(s)
            for i in range(len(word)):
                if word[i] == guess:
                    guessed_word[i] = guess
            print(' '.join(guessed_word))

            # Check if the entire word has been guessed
            if '_' not in guessed_word:
                print("Congratulations! You guessed the word:", word)
                break
        else:
            print("Incorrect!")
            attempts -= 1
            print("Attempts remaining:", attempts)

    else:
        print("Sorry, you ran out of attempts. The word was:", word)

# Call the hangman function to start the game
hangman()
```

```python
def grocery_budget_tracker(budget):
    total_spent = 0

    while True:
        # Display current budget and total spent
        print("\nCurrent budget: ₹{}".format(budget))
        print("Total spent: ₹{}".format(total_spent))

        # Ask the user for input
        user_input = input("Enter the amount spent (press 'q' to quit): ")

        # Check if the user wants to quit
        if user_input.lower() == 'q':
            break

        # Check if the input is a valid number
        if not user_input.isdigit():
            print("Invalid input. Please enter a valid number.")
            continue

        amount_spent = int(user_input)

        # Check if the amount spent exceeds the budget
        if amount_spent > budget:
            print("Warning: The amount spent exceeds the budget!")
        else:
            total_spent += amount_spent
            budget -= amount_spent

    # Display final results
    print("\nFinal budget: ₹{}".format(budget))
    print("Total spent: ₹{}".format(total_spent))

    # Check if the user stayed within budget
    if total_spent >= budget:
        print("Congratulations! You stayed within your budget.")
    else:
        print("Oops! You exceeded your budget.")

# Example usage:
budget = float(input("Enter your grocery budget: ₹"))
grocery_budget_tracker(budget)
```

```python
def check_password_strength(password):
    # Criteria for password strength
    length_criteria = 8
    uppercase_criteria = False
    lowercase_criteria = False
    digit_criteria = False
    special_character_criteria = False

    # Check length
    if len(password) >= length_criteria:
        length_criteria_met = True
    else:
        length_criteria_met = False

    # Check for uppercase characters
    for char in password:
        if char.isupper():
            uppercase_criteria = True
            break

    # Check for lowercase characters
    for char in password:
        if char.islower():
            lowercase_criteria = True
            break

    # Check for digits
    for char in password:
        if char.isdigit():
            digit_criteria = True
            break

    # Check for special characters
    special_characters = "!@#$%^&*()-_=+[{]}\|;:'\",<.>/?"
    for char in password:
        if char in special_characters:
            special_character_criteria = True
            break

    # Check overall strength
    if length_criteria_met and uppercase_criteria and lowercase_criteria and digit_criteria and special_character_criteria:
        return "Strong password! Keep it up!"
    else:
        suggestions = []
        if not length_criteria_met:
            suggestions.append("Password should have at least {} characters.".format(length_criteria))
        if not uppercase_criteria:
            suggestions.append("Include at least one uppercase letter.")
        if not lowercase_criteria:
            suggestions.append("Include at least one lowercase letter.")
        if not digit_criteria:
            suggestions.append("Include at least one digit.")
        if not special_character_criteria:
            suggestions.append("Include at least one special character (!@#$%^&*()-_=+[{]}\|;:'\",<.>/?).")

        return "Weak password. Suggestions: " + ", ".join(suggestions)

# Get password input from the user
password = input("Enter your password: ")

# Check password strength
result = check_password_strength(password)
print(result)
```

```python
def main():
    # Initialize variables
    income = 0
    expenses = 0

    # Main loop
    while True:
        print("\nPersonal Budgeting App")
        print("1. Add Income")
        print("2. Add Expense")
        print("3. View Summary")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            amount = float(input("Enter income amount: "))
            income += amount
            print("Income added successfully.")

        elif choice == '2':
            amount = float(input("Enter expense amount: "))
            expenses += amount
            print("Expense added successfully.")

        elif choice == '3':
            print("\nSummary")
            print("Income:", income)
            print("Expenses:", expenses)
            print("Balance:", income - expenses)

        elif choice == '4':
            print("Exiting the program. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 4.")
main()
```

```python
def quiz():
    # Quiz questions and answers
    questions = [
        "What is the capital of India? ",
        "Who is the first prime minister of India? ",
        "What is the largest mammal in the world? "
    ]
    answers = [
        "Delhi",
        "Jawaharlal nehru",
        "Blue whale"
    ]
    feedback = [
        "Correct! Delhi is the capital of India.",
        "Correct! jawaharlal Nehru was first prime minister of India.",
        "Correct! The blue whale is the largest mammal in the world."
    ]

    # Initialize score
    score = 0

    # Ask questions
    for i in range(len(questions)):
        response = input(questions[i]).strip().capitalize()

        # Check answer
        if response == answers[i]:
            print("Correct! " + feedback[i])
            score += 1
        else:
            print("Incorrect! The correct answer is: " + answers[i])

    # Provide overall feedback
    print("\nQuiz complete!")
    print("Your score:", score, "/", len(questions))
    if score == len(questions):
        print("Congratulations! You got all the questions correct.")
    elif score <= len(questions):
        print("Not bad! You got some of the questions correct.")
    else:
        print("Keep practicing! You can do better.")

# Run the quiz
quiz()
```

```python
def currency_converter(amount, from_currency, to_currency):
    # Conversion rates (as of a certain date)
    conversion_rates = {
        "USD": 1.0,
        "EUR": 0.92,
        "GBP": 0.79,
        "JPY": 150.38,
        "INR": 82.85,
    }

    # Check if the currencies are in the conversion rates dictionary
    if from_currency not in conversion_rates or to_currency not in conversion_rates:
        return "Invalid currency"

    # Convert the amount to the desired currency
    converted_amount = amount * (conversion_rates[to_currency] / conversion_rates[from_currency])
    return converted_amount

# Get user input
amount = float(input("Enter amount to convert: "))
from_currency = input("Enter currency to convert from (e.g., USD): ").upper()
to_currency = input("Enter currency to convert to (e.g., EUR): ").upper()

# Perform currency conversion
result = currency_converter(amount, from_currency, to_currency)
if type(result) == str:
    print(result)
else:
    print("Converted amount:", result, to_currency)
```

```python
def matrix_multiplication(matrix1, matrix2):
    result_matrix = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix2[0])):
            element = 0
            for k in range(len(matrix2)):
                element += matrix1[i][k] * matrix2[k][j]
            row.append(element)
        result_matrix.append(row)
    return result_matrix

# Define matrices
matrix_a = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix_b = [[7, 8], [9, 10], [11, 12]]

# Perform matrix multiplication
result_matrix = matrix_multiplication(matrix_a, matrix_b)

# Print the result
print("Result of matrix multiplication:")
for row in result_matrix:
    print(row)
```
